```python
def get_whole_dice_manifest(target, min_dice, spec):
    dice_map = map(lambda v: (v, dict(get_score_probability_map(target, v, spec)) ), range(min_dice, 11))
    ddm = dict(dice_map)
    plots = map(lambda v: (v, get_cumulative_probability_for_dictionary(ddm[v])), ddm.keys())
    dp = dict(plots)
    return dp
```

```python
dp = get_whole_dice_manifest(6, 1, False)
# dp2 = get_whole_dice_manifest(6, 5, True)
```

```python
from pylab import *
import matplotlib.pyplot as plt


f, axarr = plt.subplots(1, sharex=True, figsize=(10,10))
f.set_size_inches(10,10)
axarr.set_title('5-10, No Spec')
# axarr[1].set_title('5-10, With Spec')

for val in dp:
    tmp = dp[val]
    x,y = zip(*tmp)
    axarr.plot(x, y , label=val)
    
# for val in dp2:
#     tmp = dp2[val]
#     x,y = zip(*tmp)
#     axarr[1].plot(x, y , label=val)

plt.ylabel('probability')

plt.ylim(-0.05,1.05)
plt.xlim(-1.05,12.05)
plt.legend()
plt.show()
```
